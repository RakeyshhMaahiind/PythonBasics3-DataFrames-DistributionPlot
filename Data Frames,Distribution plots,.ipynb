# Handling with null values

```python
# Replace,retain or remove
import pandas as pd
import numpy as np

df1=pd.DataFrame([[1,2,3,5],[3,4,5]], columns=['exam1','exam2','exam3','exam4'])
print(df1)
```

```python
dframe=pd.DataFrame([[1,2,3],[np.nan,5,6],[7,np.nan,9],[np.nan,np.nan,np.nan]],columns=['col1','col2','col3'])
print(dframe)
```

```python
dframe.sum()
```

```python
dframe.isnull().sum()
```

```python
print(dframe)
```

```python
# dropna to remove column,inplace= true then permanent changes

dframe.dropna()
```

```python
dframe
```

```python
# if axis=0 then row wise dropping, and axis=1 then column wise dropping
dframe.dropna(how='all',axis=0)
```

```python
dframe.dropna(axis=1)
```

```python
# fill Na
dframe.fillna(0) # if inplace = false default then chnages will be not permanantly
```

```python
dframe.fillna('Not found')
```
